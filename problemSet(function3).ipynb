```python
def print_reverse(word):
    return word[::-1]

print(print_reverse("python"))
```

```python
def print_score(score_list):
    avg = sum(score_list) / len(score_list)
    return avg

print(print_score([1, 2, 3]))
```

```python
def print_even(num_list):
    for i in num_list:
        if i % 2 == 0:
            return i
        
print(print_even([1, 3, 2, 10, 12, 11, 15]))
#  ??
```
