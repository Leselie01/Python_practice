```python
def pickup_even(any_list):
    result = []
    for i in any_list:
        if i % 2 == 0:
            result.append(i)
    return result
        
print(pickup_even([3, 4, 5, 6, 7, 8]))
```

```python
def convert_int(numbers):
    for i in numbers:
        return int(numbers.replace(",", ""))
print(convert_int("1,234,567"))
```
