```python
def print_upper_price(price):
    upper_price = price * 1.3
    return upper_price

print(print_upper_price(1000))

#여기에서는 int(price) 하지 않아도 결과값이 출력됨. 왜!
```

```python
def print_arithmetic_operation(a, b):
    add = a + b
    subtract = a - b
    multiply = a * b
    divide = a / b
    list_1 = [add, subtract, multiply, divide]
    for i in list_1:
        print(i)

print_arithmetic_operation(3, 4)
# 3 + 4 = 7과 같은 형식은 어떻게 출력?
```

```python
def print_arithmetic_operation(a, b):
    add = a + b
    subtract = a - b
    multiply = a * b
    divide = a / b
    print(a, "+", b, "=", add)
    print(a, "-", b, "=", subtract)
    print(a, "*", b, "=", multiply)
    print(a, "/", b, "=", divide)
    
print_arithmetic_operation(3, 4)
```

```python
def print_max(a, b, c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    elif c > a and c > b:
        return c
    
print(print_max( 8, 3, 9))
```
